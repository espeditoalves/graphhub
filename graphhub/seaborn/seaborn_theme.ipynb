# 3. seaborn.set_style()
---

```python
seaborn.set_style(style=None, rc=None)
```
Call the function with the name of a seaborn style to set the default for all plots:

````python
sns.set_style("whitegrid")
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
````

````python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(x=["A", "B", "C"], y=[1, 3, 2])
````

A diferena entre `sns.set_theme()` e `sns.set_style()` no Seaborn  principalmente em termos de escopo e funcionalidade:

- `sns.set_theme()`:  uma funo mais abrangente que configura o tema visual global para os grficos. Ela no apenas define o estilo dos eixos, mas tambm configura outros aspectos visuais, como a paleta de cores, o contexto de plotagem (tamanho dos elementos do grfico) e a fonte.  a maneira recomendada de configurar o estilo visual de forma consistente em todas as suas visualizaes.
- `sns.set_style()`: Esta funo  mais especfica e se concentra apenas no estilo dos eixos dos grficos. Ela permite alterar elementos como a cor de fundo, se as linhas da grade esto visveis e outros detalhes estticos dos eixos. No afeta outros aspectos visuais como a paleta de cores ou o contexto de plotagem.
  
Em resumo, `sns.set_theme()`  uma funo de configurao de nvel superior que pode definir o estilo dos eixos e outras propriedades visuais, enquanto `sns.set_style()`  uma funo que se concentra especificamente no estilo dos eixos. Se voc est procurando uma configurao rpida e global, `sns.set_theme()`  a escolha ideal. Para ajustes mais finos relacionados apenas aos eixos, `sns.set_style()` seria mais apropriado.

# Configurao especifica para cada subplot usando Matplotlib

````python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

# Gerando dados fictcios
np.random.seed(0)
dados = {
    'sua_coluna': np.random.randn(100),
    'sua_coluna_x': np.random.rand(100),
    'sua_coluna_y': np.random.rand(100) * 100,
    'categoria': np.random.choice(['A', 'B', 'C', 'D'], 100)
}

# Criando o DataFrame
seu_dataframe = pd.DataFrame(dados)

# Exibindo as primeiras linhas do DataFrame
seu_dataframe.head()

````

https://matplotlib.org/stable/api/axes_api.html

````python
# Criando uma figura com uma grade de subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Plotando dados usando Seaborn em cada subplot especfico
sns.histplot(data=seu_dataframe, x="sua_coluna", ax=axes[0, 0])
sns.boxplot(data=seu_dataframe, x="sua_coluna", ax=axes[0, 1])
sns.scatterplot(data=seu_dataframe, x="sua_coluna_x", y="sua_coluna_y", ax=axes[1, 0])
sns.lineplot(data=seu_dataframe, x="sua_coluna_x", y="sua_coluna_y", ax=axes[1, 1])

# Personalizando o subplot axes[0, 0]
axes[0, 0].set_title('Histograma')
axes[0, 0].set_xlabel('Eixo X')
axes[0, 0].set_ylabel('Eixo Y')
axes[0, 0].grid(False)  # Removendo a grade
axes[0, 0].tick_params(axis='x', which='both', bottom=False, top=False)  # Removendo ticks do eixo X
axes[0, 0].tick_params(axis='y', which='both', left=False, right=False)  # Removendo ticks do eixo Y

# Personalizando o subplot axes[0, 1]
axes[0, 1].set_title('Boxplot')
axes[0, 1].set_xlabel('Eixo X')
axes[0, 1].set_ylabel('Eixo Y')
axes[0, 1].spines['top'].set_visible(False)  # Removendo a borda superior
axes[0, 1].spines['right'].set_visible(False)  # Removendo a borda direita
axes[0, 1].set_facecolor('#f0f0f0')  # Mudando a cor de fundo

# Personalizando o subplot axes[1, 0]
axes[1, 0].set_title('Scatterplot')
axes[1, 0].set_xlabel('Eixo X')
axes[1, 0].set_ylabel('Eixo Y')
axes[1, 0].spines['left'].set_color('red')  # Mudando a cor da borda esquerda
axes[1, 0].spines['bottom'].set_linestyle('--')  # Mudando o estilo da borda inferior

# Personalizando o subplot axes[1, 1]
axes[1, 1].set_title('Grfico de Linha')
axes[1, 1].set_xlabel('Eixo X')
axes[1, 1].set_ylabel('Eixo Y')
axes[1, 1].spines['left'].set_linewidth(2)  # Aumentando a largura da borda esquerda
axes[1, 1].spines['bottom'].set_linewidth(2)  # Aumentando a largura da borda inferior

# Ajustando o layout para evitar sobreposio
plt.tight_layout()

# Mostrando o grfico
plt.show()
````

# Classe Spine do Matplotlib

A classe `matplotlib.spines.Spine`  utilizada para criar "espinhas", que so as linhas que conectam as marcas de escala dos eixos e delimitam os limites da rea de dados do grfico.

## Parmetros e Funcionalidades

- **axes**: O objeto `Axes` que contm a espinha. Cada eixo pode ter vrias espinhas que definem os limites da rea de plotagem.
- **spine_type**: Uma string que especifica o tipo de espinha. Os tipos comuns so 'left', 'right', 'top' e 'bottom', que correspondem aos lados do eixo.
- **path**: Um objeto `Path` que define a forma da espinha. Geralmente,  uma linha reta, mas pode ser qualquer forma geomtrica.
- **kwargs**: Argumentos de palavra-chave adicionais que so passados para a classe base `Patch`. Isso inclui propriedades como cor, largura da linha, estilo da linha e outros atributos visuais.

As espinhas so subclasses de `Patch`, o que significa que elas herdam muitos comportamentos de patches, que so formas bsicas desenhveis no Matplotlib. As espinhas podem desenhar uma linha, um crculo ou um arco, dependendo se os mtodos `set_patch_line`, `set_patch_circle` ou `set_patch_arc` foram chamados. O padro  linha.

## Mtodos Comuns

- **set_position**: Define a posio da espinha em relao ao eixo.
- **set_color**: Altera a cor da espinha.
- **set_linewidth**: Define a largura da linha da espinha.
- **set_visible**: Controla a visibilidade da espinha.



````python
# Criando dados fictcios
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando os dados
ax.plot(x, y)

# Acessando as espinhas e personalizando-as
spines = ax.spines
spines['right'].set_visible(False)  # Removendo a espinha direita
spines['top'].set_visible(False)    # Removendo a espinha superior
spines['left'].set_color('red')     # Mudando a cor da espinha esquerda
spines['bottom'].set_linestyle('--')# Mudando o estilo da espinha inferior

# Exibindo o grfico
plt.show()
````
