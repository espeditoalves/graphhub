# 3. seaborn.set_style()
---

```python
seaborn.set_style(style=None, rc=None)
```
Call the function with the name of a seaborn style to set the default for all plots:

````python
sns.set_style("whitegrid")
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])
````

````python
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.lineplot(x=["A", "B", "C"], y=[1, 3, 2])
````

A diferença entre `sns.set_theme()` e `sns.set_style()` no Seaborn é principalmente em termos de escopo e funcionalidade:

- `sns.set_theme()`: É uma função mais abrangente que configura o tema visual global para os gráficos. Ela não apenas define o estilo dos eixos, mas também configura outros aspectos visuais, como a paleta de cores, o contexto de plotagem (tamanho dos elementos do gráfico) e a fonte. É a maneira recomendada de configurar o estilo visual de forma consistente em todas as suas visualizações.
- `sns.set_style()`: Esta função é mais específica e se concentra apenas no estilo dos eixos dos gráficos. Ela permite alterar elementos como a cor de fundo, se as linhas da grade estão visíveis e outros detalhes estéticos dos eixos. Não afeta outros aspectos visuais como a paleta de cores ou o contexto de plotagem.
  
Em resumo, `sns.set_theme()` é uma função de configuração de nível superior que pode definir o estilo dos eixos e outras propriedades visuais, enquanto `sns.set_style()` é uma função que se concentra especificamente no estilo dos eixos. Se você está procurando uma configuração rápida e global, `sns.set_theme()` é a escolha ideal. Para ajustes mais finos relacionados apenas aos eixos, `sns.set_style()` seria mais apropriado.

# Configuração especifica para cada subplot usando Matplotlib

````python
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import numpy as np

# Gerando dados fictícios
np.random.seed(0)
dados = {
    'sua_coluna': np.random.randn(100),
    'sua_coluna_x': np.random.rand(100),
    'sua_coluna_y': np.random.rand(100) * 100,
    'categoria': np.random.choice(['A', 'B', 'C', 'D'], 100)
}

# Criando o DataFrame
seu_dataframe = pd.DataFrame(dados)

# Exibindo as primeiras linhas do DataFrame
seu_dataframe.head()

````

https://matplotlib.org/stable/api/axes_api.html

````python
# Criando uma figura com uma grade de subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Plotando dados usando Seaborn em cada subplot específico
sns.histplot(data=seu_dataframe, x="sua_coluna", ax=axes[0, 0])
sns.boxplot(data=seu_dataframe, x="sua_coluna", ax=axes[0, 1])
sns.scatterplot(data=seu_dataframe, x="sua_coluna_x", y="sua_coluna_y", ax=axes[1, 0])
sns.lineplot(data=seu_dataframe, x="sua_coluna_x", y="sua_coluna_y", ax=axes[1, 1])

# Personalizando o subplot axes[0, 0]
axes[0, 0].set_title('Histograma')
axes[0, 0].set_xlabel('Eixo X')
axes[0, 0].set_ylabel('Eixo Y')
axes[0, 0].grid(False)  # Removendo a grade
axes[0, 0].tick_params(axis='x', which='both', bottom=False, top=False)  # Removendo ticks do eixo X
axes[0, 0].tick_params(axis='y', which='both', left=False, right=False)  # Removendo ticks do eixo Y

# Personalizando o subplot axes[0, 1]
axes[0, 1].set_title('Boxplot')
axes[0, 1].set_xlabel('Eixo X')
axes[0, 1].set_ylabel('Eixo Y')
axes[0, 1].spines['top'].set_visible(False)  # Removendo a borda superior
axes[0, 1].spines['right'].set_visible(False)  # Removendo a borda direita
axes[0, 1].set_facecolor('#f0f0f0')  # Mudando a cor de fundo

# Personalizando o subplot axes[1, 0]
axes[1, 0].set_title('Scatterplot')
axes[1, 0].set_xlabel('Eixo X')
axes[1, 0].set_ylabel('Eixo Y')
axes[1, 0].spines['left'].set_color('red')  # Mudando a cor da borda esquerda
axes[1, 0].spines['bottom'].set_linestyle('--')  # Mudando o estilo da borda inferior

# Personalizando o subplot axes[1, 1]
axes[1, 1].set_title('Gráfico de Linha')
axes[1, 1].set_xlabel('Eixo X')
axes[1, 1].set_ylabel('Eixo Y')
axes[1, 1].spines['left'].set_linewidth(2)  # Aumentando a largura da borda esquerda
axes[1, 1].spines['bottom'].set_linewidth(2)  # Aumentando a largura da borda inferior

# Ajustando o layout para evitar sobreposição
plt.tight_layout()

# Mostrando o gráfico
plt.show()
````

# Classe Spine do Matplotlib

A classe `matplotlib.spines.Spine` é utilizada para criar "espinhas", que são as linhas que conectam as marcas de escala dos eixos e delimitam os limites da área de dados do gráfico.

## Parâmetros e Funcionalidades

- **axes**: O objeto `Axes` que contém a espinha. Cada eixo pode ter várias espinhas que definem os limites da área de plotagem.
- **spine_type**: Uma string que especifica o tipo de espinha. Os tipos comuns são 'left', 'right', 'top' e 'bottom', que correspondem aos lados do eixo.
- **path**: Um objeto `Path` que define a forma da espinha. Geralmente, é uma linha reta, mas pode ser qualquer forma geométrica.
- **kwargs**: Argumentos de palavra-chave adicionais que são passados para a classe base `Patch`. Isso inclui propriedades como cor, largura da linha, estilo da linha e outros atributos visuais.

As espinhas são subclasses de `Patch`, o que significa que elas herdam muitos comportamentos de patches, que são formas básicas desenháveis no Matplotlib. As espinhas podem desenhar uma linha, um círculo ou um arco, dependendo se os métodos `set_patch_line`, `set_patch_circle` ou `set_patch_arc` foram chamados. O padrão é linha.

## Métodos Comuns

- **set_position**: Define a posição da espinha em relação ao eixo.
- **set_color**: Altera a cor da espinha.
- **set_linewidth**: Define a largura da linha da espinha.
- **set_visible**: Controla a visibilidade da espinha.



````python
# Criando dados fictícios
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Criando a figura e os eixos
fig, ax = plt.subplots()

# Plotando os dados
ax.plot(x, y)

# Acessando as espinhas e personalizando-as
spines = ax.spines
spines['right'].set_visible(False)  # Removendo a espinha direita
spines['top'].set_visible(False)    # Removendo a espinha superior
spines['left'].set_color('red')     # Mudando a cor da espinha esquerda
spines['bottom'].set_linestyle('--')# Mudando o estilo da espinha inferior

# Exibindo o gráfico
plt.show()
````
