````python
import matplotlib.pyplot as plt
import numpy as np
````

## Exemplo 1: Uso simples com matplotlib.pyplot

Este exemplo usa o módulo `pyplot` para criar um gráfico de linha simples de forma rápida e direta:

````python
# Dados para plotagem
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Criação rápida de um gráfico de linha
plt.plot(x, y)
plt.title('Gráfico de Linha Simples')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
````

# Exemplo 2: Uso avançado com a interface orientada a objetos

Este exemplo utiliza a interface orientada a objetos para criar uma figura com dois subplots, cada um com sua própria personalização:


````python
# Dados para plotagem
x = np.linspace(0, 2, 100)
y1 = np.sin(2 * np.pi * x)
y2 = np.cos(2 * np.pi * x)

# Criação da figura e dos eixos
fig, (ax1, ax2) = plt.subplots(2, 1)  # 2 linhas, 1 coluna

# Primeiro subplot
ax1.plot(x, y1, 'r-')  # Linha vermelha
ax1.set_title('Seno')
ax1.set_ylabel('Amplitude')

# Segundo subplot
ax2.plot(x, y2, 'b--')  # Linha tracejada azul
ax2.set_title('Cosseno')
ax2.set_xlabel('Tempo (s)')
ax2.set_ylabel('Amplitude')

# Ajuste automático do layout para evitar sobreposições
plt.tight_layout()
plt.show()
````

````python
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
plt.show()
````

# Algumas funções e jeitos de usar

### Função plot() - Gráficos de Linha


https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

- **x, y**: Coordenadas horizontais e verticais dos pontos de dados. Se apenas `y` for fornecido, `x` será automaticamente gerado como uma sequência de 0 a N-1.
- **fmt**: Uma string de formatação opcional para definir a cor, marcador e estilo de linha do gráfico.
- **color**: A cor da linha. Pode ser um nome de cor ou um código de cor hexadecimal.
- **linestyle**: O estilo da linha, como sólido (`'-'`), pontilhado (`'--'`), tracejado (`':'`), etc.
- **linewidth**: A espessura da linha.
- **marker**: O estilo do marcador, como círculo (`'o'`), quadrado (`'s'`), estrela (`'*'`), etc.
- **label**: A etiqueta para a legenda do gráfico.

Além desses, existem muitos outros argumentos que podem ser usados para controlar aspectos como:
- **alpha**: A transparência da linha.
- **markerfacecolor**: A cor de preenchimento do marcador.
- **markeredgecolor**: A cor da borda do marcador.


No Matplotlib, o argumento `fmt` é uma string de formatação compacta que especifica a cor, o marcador e o estilo da linha, 
e não deve ser passado como um argumento de palavra-chave nomeado. Em vez disso, ele é usado como parte dos argumentos posicionais.

Neste código corrigido, `co--` é a string de formatação onde ‘c’ representa a cor ciano, ‘o’ indica marcadores de círculo, e ‘–’ define o estilo da linha como pontilhado. Não é necessário usar `x=` ou `y=` quando se usa `fmt`.
```
O Formato da string deve ser:
fmt = '[marker][line][color]'
```

````python
# Dados para plotagem
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Cria uma figura e um conjunto de subplots
fig, ax = plt.subplots()

# Plotagem no subplot
ax.plot(x, y, 'co--', label='linha ciano pontilhada com marcadores de círculo')

# Personalizações adicionais
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico Simples com fmt')
ax.legend()
ax.grid(True)

# Mostra a figura
plt.show()

````

Neste exemplo, `co--` é a string de formatação onde:

- `c` representa a cor ciano.
- `o` indica que estamos usando marcadores de círculo.
- `--` define o estilo de linha como pontilhado.

### Função set()
A função `.set()` é um método dos objetos de eixo (`Axes`), que são obtidos, por exemplo, através de `fig, ax = plt.subplots()`.

````python
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
# Outra opção
# ax.set_title('About as simple as it gets, folks')
# ax.set_xlabel('time (s)')
# ax.set_ylabel('voltage (mV)')
ax.grid()
plt.show()
````
